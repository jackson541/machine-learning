Existem formas mais avançadas para se capturar dados específicos da tabela utilizando iloc e loc.

O iloc é utilizado com base no index:

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.iloc.html


```python
# selecionar a primeira linha do DataFrame
wine_data.iloc[0]
```

eles receber diversos tipos de parâmetros que se comportam como listas (com números ou booleanos) ou até mesmo funções.



```python
# o segundo parâmetro são as colunas que devem ser exibidas.
# para listar todos os dados de uma coluna:

wine_data.iloc[:, 0]
```

Já o loc recebe parâmetros de forma bem parecida com o iloc, com a diferença de que esses parâmetros são labels ou booleanos

### Condicionais

A manipulação condicional permite ter muito mais controle sobre os dados da tabela e podem ser feitos em conjunto com o loc e o iloc


```python
# a condicional irá retornar um booleano como é esperado 
wine_data['country'] == 'Italy'
```

```python
# filtrando vinhos por país específico

wine_data.loc[wine_data['country'] == 'Italy']
```

```python
# filtrando por país e pontuação do vinho

wine_data.loc[(wine_data['country'] == 'Italy') & (wine_data['points'] >= 90)]
```

Existem muitos métodos do pandas que ajudam a realizar consultas condicionais, um deles é o `isin` que verifica se os elementos do DataFrame estão contidos em uma lista.

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.isin.html

```python
# filtrando por vinhos da Itália ou da França

wine_data.loc[wine_data['country'].isin(['Italy', 'France'])]
```

### Iteração sobre o DataSet
É possível fazer iterações sobre um DataSet de forma muito simples, seja para extrair ou para modificar dados

```python
# a função map permite iterar sobre uma coluna fazendo modificações nela e retornando essas modificações
# ela recebe uma função que irá fazer as alterações

# nesse exemplo estamos vendo a diferença de cada valor da coluna points em relação a média
points_mean = wine_data['points'].mean()
wine_data['points'].map(lambda point: point - points_mean)
```

```python
# uma outra forma de iterar é utilizando a função apply, a diferença entre ela e o map é que o map recebe apenas um
# único valor, enquanto ela recebe toda uma linha, ou coluna, para iterar sobre

import numpy as np

def sum_price_and_points(row):
    row['price'] = row['price'] + row['points']
    return row
    
wine_data.apply(sum_price_and_points, axis='columns')
```

```python
# vale lembrar que o map e o apply não alteram os dados do DataFrame

wine_data.head()
```

```python
# também é possível fazer a iteração sem utilizar funções
wine_data['points'] + wine_data['price']
```

A operação mostrada acima é mais rápida do que quando utilizado o `map` ou `apply` com funções. No entanto, essas 2 funções de iteração nos dão mais possibilidades de condicionais dentro das funções e mais personalização

   ### Agrupar e ordenar
  `groupby` é a função que permite agrupar as linhas, ou colunas, a partir de algo, seja o nome de uma coluna, função, mapping, etc.

```python
# uma forma de reescrever a função values_count utilizando o agrupamento
wine_data.groupby('points').count()
```

```python
# é muito útil para pegar dados específicos a partir do agrupamento
# exemplo para pegar o preço médio dos vinhos por pontuação:

wine_data.groupby('points')['price'].mean()
```

```python
# cada grupo pode ser tratado como um DataFrame isolado, assim podemos aplicar funções sobre cada um deles

# no exemplo está pegando o título do primeiro vinho de cada adega:

wine_data.groupby('winery').apply(lambda df: df['title'].iloc[0])
```

```python
# também é possível fazer o agrupamento utilizando mais de uma coluna

# no exemplo abaixo, estamos agrupando primeiro por país e depois por província e pegando o vinho com a 
# maior pontuação de cada província:

wine_data.groupby(['country', 'province']).apply(lambda df: df.loc[df['points'].idxmax()])
```

```python
# uma outra função utilizada junto com o agrupamento é a agregação (agg), que nos permite rodar diversas funções
# em paralelo em um mesmo DataFrame

# no exemplo abaixo, estamos aplicando 3 funções no preço do agrupamento por país

wine_data.groupby('country')['price'].agg([len, min, max])
```

Como é possível notar acima, no exemplo do melhor vinho das províncias, o pandas também consegue trabalhar com DataFrames que contém múltiplos índices.

```python
# esse é mais um exemplo de múltiplos índices:

multi_indexes = wine_data.groupby(['country', 'province'])['description'].agg([len])
multi_indexes
```

```python
# também é possível voltar para a estrutura de um único índice:

multi_indexes.reset_index()
```

```python
# algo que pode ser útil para visualizar os dados é utilizar ordenação:

wine_data.sort_values(by='price')
```

```python
# também é possível ordenar por mais um atributo (seja para coluna ou linha)

new_order = multi_indexes.sort_values(['country', 'len'])
new_order
```

```python
# e existe um método para caso seja necessário voltar a ordenação por index

new_order.reset_index()
```

### Tipos de dados e valores nulos

```python
# para verificar os tipos de dados de cada coluna, é possível utilizar o método dtypes:

wine_data.dtypes
```

```python
# é possível realizar a transformação do tipo de uma coluna para outro
# no exemplo abaixo, está transformando a coluna de pontos de int para float:

wine_data['points'].astype('float64')
```

Para identificar valores nulos em uma tabela, ou coluna, é possível utilizar os métodos `isnull` ou `notnull` (para o caso de pegar apenas os não nulos)

```python
# pegar todas as linhas em que o nome do país se encontra vazio:

wine_data[pd.isnull(wine_data['country'])]
```

Existem algumas estratégias para lidar com esses valores nulos, como preencher com a média (em caso de números), com algum valor que tenha aparecido previamente ou com alguma outra identificação (a exemplo de "Unknown" para strings). As vezes pode ser viável apenas excluir essas linhas da tabela quando elas se encontram em pequena quantidade.

Essas substituições podem ser feitas através do método `fillna`

```python
# substituindo o nomes da region_2 que estão vazios por uma outra palavra:

wine_data['region_2'].fillna('Unknown')
```

```python
# também é possível alterar valores da tabela que estejam preenchidos com o método replace:

wine_data['taster_twitter_handle'].replace("@kerinokeefe", "@kerino")
```
